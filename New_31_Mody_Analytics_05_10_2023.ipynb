```python
import os
import openpyxl
import csv

# Specify the folder containing the Excel files
folder_path = r'C:/Users/GenepoweRx_Madhu/Downloads/MODY_new_05_10_2023/'

# Specify the output folder for TSV files
output_folder = r'C:/Users/GenepoweRx_Madhu/Downloads/MODY_new_05_10_2023/TSV/'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Specify the sheet name you want to extract (modify this as needed)
sheet_name = '(Variants 101,584) Filter Vari'

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx'):
        # Construct the full path to the Excel file
        excel_file_path = os.path.join(folder_path, filename)

        # Open the Excel workbook
        workbook = openpyxl.load_workbook(excel_file_path, read_only=True)

        try:
            # Select the desired sheet
            sheet = workbook[sheet_name]

            # Construct the output TSV file path
            tsv_file_path = os.path.join(output_folder, f'{os.path.splitext(filename)[0]}_snp.tsv')

            # Open the TSV file for writing
            with open(tsv_file_path, 'w', newline='') as tsv_file:
                tsv_writer = csv.writer(tsv_file, delimiter='\t')

                # Iterate through rows in the sheet, excluding the header (first row), and write them to the TSV file
                is_header = True
                for row in sheet.iter_rows(values_only=True):
                    if is_header:
                        is_header = False
                    else:
                        tsv_writer.writerow(row)

            print(f"Sheet '{sheet_name}' from '{filename}' has been extracted (excluding header) and saved as '{tsv_file_path}'.")
        except KeyError:
            print(f"Sheet '{sheet_name}' not found in '{filename}'.")

print("Extraction process completed for all Excel files in the folder.")
```

```python
import pandas as pd
import os

# Set the path to the folder containing your Excel files
folder_path = r'C:/Users/GenepoweRx_Madhu/Downloads/MODY_new_05_10_2023/'

# List all Excel files in the folder
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Iterate through each Excel file
for excel_file in excel_files:
    # Define the Excel file path
    excel_file_path = os.path.join(folder_path, excel_file)

    # Read the Excel file using pandas without loading the entire file
    xls = pd.ExcelFile(excel_file_path)

    # Iterate through each sheet in the Excel file
    for sheet_name in xls.sheet_names:
        # Create a DataFrame for the current sheet
        df = pd.read_excel(excel_file_path, sheet_name=sheet_name)

        # Define the TSV file path (you can change the extension to .tsv)
        tsv_file_path = os.path.join(folder_path, f'{excel_file}_{sheet_name}.tsv')

        # Write the DataFrame to a TSV file without loading it entirely into memory
        df.to_csv(tsv_file_path, sep='\t', index=False)

    # Close the Excel file
    xls.close()

print("Conversion complete.")
```
