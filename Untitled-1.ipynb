```python
def get_fourth_position(row):
    elements = row.split(',')
    fourth_element = ''
    for item in elements:
        sub_elements = item.split('|')
        if len(sub_elements) >= 4:
            fourth_element += sub_elements[3] + ','
    return fourth_element.rstrip(',')

df['New_Column'] = df['CSQ'].apply(get_fourth_position)
df

def get_fourth_position(row):
    elements = row.split(',')
```
